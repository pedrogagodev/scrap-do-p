# Census API Variables Search (Modularized Version)
This notebook uses the `census_engine.py` module to download and search the 2024 ACS 1-Year variables.

```python
import pandas as pd
from census_engine import CensusEngine
```

```python
# Load data using the shared engine
df = CensusEngine.load_data()
print(f"Total variables loaded: {len(df)}")
df.head(3)
```

## 🔍 Search Interface
Change the value of `search_term` below and run the cell to search within the `Label` column.

```python
# ---------------------------------------------------------
# 1. TYPE YOUR SEARCH TERM HERE
# ---------------------------------------------------------
search_term = 'poverty'

# ---------------------------------------------------------
# 2. EXECUTE SEARCH USING ENGINE
# ---------------------------------------------------------
results = CensusEngine.search(df, search_term)

print(f"Found {len(results)} results for term: '{search_term}'\n")

# Display results without truncation
pd.set_option('display.max_colwidth', None)
results
```
